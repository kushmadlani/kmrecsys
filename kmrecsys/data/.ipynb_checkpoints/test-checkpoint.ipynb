```python
def sparse_to_list(X):
    """list of nonzero indices per row of sparse X"""
    result = np.split(X.indices, X.indptr)[1:-1]
    result = [list(r) for r in result]
    return result
```

```python
x = sparse_to_list(data['val_masked'])
y = [len(i) for i in x]
print(y)
```
